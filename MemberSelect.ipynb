This code is intended to select 12 random candidates, including 6 Republicans and 6 Democrats.

```python
import pandas as pd
data = pd.read_csv('weball24.txt', sep='|')
```

```python
democrats = data[data['CAND_PTY_AFFILIATION'] == 'DEM'].copy()
republicans = data[data['CAND_PTY_AFFILIATION'] == 'REP'].copy()
```

```python
demoMedian = democrats['TTL_RECEIPTS'].median()
repubMedian = republicans['TTL_RECEIPTS'].median()

print("Low Income Democrats:\n")
print(democrats[democrats['TTL_RECEIPTS'] < demoMedian]["CAND_NAME"].sample(n=3).to_string(index=False))

print("\nHigh Income Democrats:\n")
print(democrats[democrats['TTL_RECEIPTS'] >= demoMedian]["CAND_NAME"].sample(n=3).to_string(index=False))

print("\nLow Income Republicans:\n")
print(republicans[republicans['TTL_RECEIPTS'] < repubMedian]["CAND_NAME"].sample(n=3).to_string(index=False))

print("\nHigh Income Republicans:\n")
print(republicans[republicans['TTL_RECEIPTS'] >= repubMedian]["CAND_NAME"].sample(n=3).to_string(index=False))
```
